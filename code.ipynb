```python
import pandas as pd
```

```python
df=pd.read_csv("data.csv")
df.head()
```

```python
df["CAL BP"].min()
```

```python
df["CAL BP"].max()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df["CAL BP"], df["Delta 14C"])
plt.xlabel("Calendar Age (years BP)")
plt.ylabel("Δ14C (per mil)")
plt.title("Atmospheric Radiocarbon Variation Over Time")
plt.gca().invert_xaxis()   # recent time on right like scientific plots
plt.show()
```

```python
df["offset"] = df["14C age"] - df["CAL BP"]

plt.figure(figsize=(10,5))
plt.plot(df["CAL BP"], df["offset"])
plt.xlabel("Calendar Age (BP)")
plt.ylabel("Radiocarbon Age Error")
plt.title("Error Between Radiocarbon Age and True Age")
plt.gca().invert_xaxis()
plt.show()
```
